```python
import pandas as pd
pd.set_option('display.max_column', 250)
import numpy as np
```

```python
import arff
data = arff.load(open('../data/poker-lsn.arff'))
cnames = [i[0] for i in data['attributes']]
df = pd.DataFrame(data['data'], columns=cnames)
df.head()
```

```python
df = pd.get_dummies(df.iloc[:,1:-1]).join(df['class'])
```

```python
df.head()
```

```python
df.to_csv('poker_onehot.csv')
```
