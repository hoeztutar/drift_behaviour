```python
def get_info(df):
    cols = ["classifications correct (percent)", 
            "Kappa Statistic (percent)", 
            "F1 Score (percent)", 
            "Precision (percent)", 
            "Recall (percent)",
            "Change detected"
           ]

    print("Average accuracy: {:.2f}%".format(df[cols[0]].mean()))
    print("Average Kappa Statistic: {:.2f}%".format(df[cols[1]].mean()))
    print("Average F1 Score: {:.2f}%".format(df[cols[2]].mean()))
    print("Average Precision: {:.2f}%".format(df[cols[3]].mean()))
    print("Average Recall: {:.2f}%".format(df[cols[4]].mean()))
    print("Total change detected: {:.0f}".format(df[cols[5]].sum()))

    for col in cols[:-1]:
        df[col].plot()
    plt.xlabel('Instances')
    plt.ylabel('Percent')
    plt.xticks([])
    plt.legend(loc=3)
    plt.show()
```

```python
get_info(df)
```
